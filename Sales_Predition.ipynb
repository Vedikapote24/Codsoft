```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('advertising.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.hist(bins=20, figsize=(10,8))
plt.suptitle("Distribution of Advertising Budgets and Sales")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Advertising Channels and Sales")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='TV', y='Sales', data=df)
plt.title("TV Advertising vs Sales")
plt.show()
```

```python
sns.regplot(x='TV', y='Sales', data=df)
plt.title("TV Advertising Impact on Sales")
plt.show()
```

```python
sns.regplot(x='Radio', y='Sales', data=df)
plt.title("Radio Advertising Impact on Sales")
plt.show()
```

```python
sns.regplot(x='Newspaper', y='Sales', data=df)
plt.title("Newspaper Advertising Impact on Sales")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title("Boxplot of Advertising Budget and Sales")
plt.show()
```

```python
df_sorted = df.sort_values(by='Sales')

plt.figure(figsize=(8,4))
plt.plot(df_sorted['Sales'].values)
plt.title("Sales Trend Across Observations")
plt.ylabel("Sales")
plt.show()
```

```python
avg_spend = df[['TV','Radio','Newspaper']].mean()

avg_spend.plot(kind='bar', figsize=(6,4))
plt.title("Average Advertising Spend by Channel")
plt.ylabel("Average Budget")
plt.show()
```
